Replace invalid 0 values with mean

```python
# Replace invalid 0 values with mean
cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

for col in cols:
    mean_value = df[df[col] != 0][col].mean()
    df[col] = df[col].replace(0, round(mean_value, 2))

df.head()
```

Histogram for Numerical Columns

```python
num_cols = ["Pregnancies", "Glucose", "BloodPressure"]

plt.figure(figsize=(10,6))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], bins=20)
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Diabetes Outcome Countplot

```python
sns.countplot(x = "Pregnancies", hue="Outcome", data=df)
plt.title("Diabetes Outcome Distribution")
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Glucose", y="BMI", hue="Outcome", data=df)
plt.title("Glucose vs BMI")
plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="Reds")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Outcome", y="BMI", hue="Outcome", data=df, palette = ["pink","red"])

plt.title("Average BMI by Outcome")
plt.xlabel("Outcome")
plt.ylabel("Average BMI")
plt.legend(title="Outcome")

plt.show()
```

```python
# Check the number of duplicate rows
print("Duplicate Rows:", df.duplicated().sum())

# Remove duplicate rows
df = df.drop_duplicates()

# Verify that duplicates have been removed
print("Duplicate Rows After Removal:", df.duplicated().sum())

# Display the new shape of the dataset
print("Dataset Shape After Removing Duplicates:", df.shape)
```

```python
# One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=["Outcome"], drop_first=True)

# Display the encoded dataset
print(df_encoded.head())
print("\nShape:", df_encoded.shape)
```
